```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("Mean")
print(df.mean())

print("\nMedian")
print(df.median())

print("\nMode")
print(df.mode().iloc[0])
```

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("Range")
print(df.max() - df.min())

print("\nVariance")
print(df.var())

print("\nStandard Deviation")
print(df.std())

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

print("\nIQR")
print(IQR)
```

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("Skewness")
print(df.skew())

print("\nKurtosis")
print(df.kurt())
```
