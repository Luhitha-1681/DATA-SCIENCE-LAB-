```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

sns.histplot(
    df['sepal length (cm)'],
    bins=20,
    kde=True
)

plt.show()
```

```python
sns.boxplot(
    x=df['sepal length (cm)']
)

plt.show()
```

```python
species_counts = pd.Series(
    iris.target
).value_counts()

plt.pie(
    species_counts,
    labels=iris.target_names,
    autopct='%1.1f%%'
)

plt.show()
```

```python
sns.scatterplot(
    x=df['sepal length (cm)'],
    y=df['sepal width (cm)']
)

plt.show()
```

```python
plt.plot(
    df['sepal length (cm)']
)

plt.show()
```

```python
df['species'] = iris.target

sns.barplot(
    x='species',
    y='sepal length (cm)',
    data=df
)

plt.show()
```

```python
sns.heatmap(
    df.iloc[:, :4].corr(),
    annot=True
)

plt.show()
```

```python
plt.scatter(
    df['sepal length (cm)'],
    df['sepal width (cm)'],
    s=df['petal length (cm)'] * 20,
    alpha=0.5
)

plt.show()
```

```python
import seaborn as sns

sns.pairplot(
    df.iloc[:, :4]
)

plt.show()
```

```python
sns.violinplot(
    x='species',
    y='sepal length (cm)',
    data=df
)

plt.show()
```
